```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'aluno_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'idade': [18, 22, 35, 19, 28, 45, 21, 30],
    'nota_media': [8.5, 4.2, 7.0, 3.1, 9.0, 6.5, 2.8, 7.8],
    'faltas': [2, 20, 5, 30, 1, 8, 35, 4],
    'evadiu': [0, 1, 0, 1, 0, 0, 1, 0]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
scaler = MinMaxScaler()
```

```python
df_minmax = df.drop(['aluno_id','evadiu'],axis=1)
```

```python
df_minmax
```

```python
df_minmax_scaled = scaler.fit_transform(df_minmax)
df_minmax_scaled = pd.DataFrame(df_minmax_scaled, columns=df_minmax.columns)
```

```python
df_minmax_scaled
```

```python
scaler_2= StandardScaler()
```

```python
df_standard = df.drop(['aluno_id','evadiu'],axis=1)
df_standard_scaled = scaler_2.fit_transform(df_standard)
df_standard_scaled = pd.DataFrame(df_standard_scaled, columns = df_standard.columns)
```

```python
df_standard_scaled
```

```python
print("=== Min-Max Scaler ===")
print(df_minmax_scaled.describe().round(2))

print("\n=== Standard Scaler ===")
print(df_standard_scaled.describe().round(2))
```
